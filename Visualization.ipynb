```python
def infoFactorDims():
    import math
    import sys
    allFactors =np.load("datasets/dsprites_full_factors.npy")
    """
    0 - shape (3 different values)
    1 - scale (6 different values)
    2 - orientation (40 different values)
    3 - position x (32 different values)
    4 - position y (32 different values)
    """
    minfos=np.zeros((allFactors.shape[1],10))
    for a in range(allFactors.shape[1]):
        factorVector=allFactors[:,a]
        for b in range(10):
            repVector=reps[:,b]
            bins=np.histogram(repVector,bins=20
#                           ,range=(-2.5,2.5)
                         )[1]
            repDiscrete=np.digitize(repVector,bins=bins)
    #         print(repDiscrete)
            minfos[a][b]=sklearn.metrics.normalized_mutual_info_score(factorVector,repDiscrete,
                                                                average_method="arithmetic"
                                                          )
    #         minfos[a][b]=sklearn.metrics.mutual_info_score(factorVector,repDiscrete)
    # minfos[:,10]=sum(minfos.T)
    plt.imshow(minfos,cmap="cividis")
    cb = plt.colorbar(aspect=10,shrink=0.62)
    plt.show()
    np.set_printoptions(threshold=sys.maxsize)

    print("minfos vertical")
    sumsVer= sum(minfos.T).reshape(-1,1)
    plt.imshow(sumsVer)
    cb = plt.colorbar()
    plt.show()
    
    print("minfos horizontal")
    sumHor= sum(minfos).reshape(1,-1)
    plt.imshow(sumHor)
    cb = plt.colorbar(orientation="horizontal")
    plt.show()

    # print(minfos)
infoFactorDims()
```
